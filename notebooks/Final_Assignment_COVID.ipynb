**Data preparation**

To analyse the data and come to any conclusion I first downloaded the COVID data set from the internet and selected all relevant columns. 

```python
import altair as alt

import pandas as pd
```

```python
covid = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv", parse_dates=['date'])

included_countries = ['NOR', 'DNK', 'SWE', 'GRL', 'FIN', 'ISL', 'FRO','ALA']

included_columns = ['iso_code','date','people_vaccinated','total_deaths','new_deaths','new_cases', 'total_cases','total_tests','new_tests']

covid = covid.loc[covid.iso_code.isin(included_countries) & (covid.date.dt.year == 2021),included_columns]
covid.head()
```

Now that I have my data frame called COVID I can see all the individual countries and the different variables. To confirm all selected coloumns I returned the label of each column in the DataFrame.
